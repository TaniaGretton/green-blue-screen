```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2

%matplotlib inline
```

```python
image = mpimg.imread('/Users/taniagretton/miniconda3/CHALLENGES/tomatito.jpg')
image.shape
```

```python
plt.imshow(image)
```

```python
lower_green = np.array([0,240,0])
upper_green = np.array([255,255,255])
mask = cv2.inRange(image,lower_green,upper_green)
plt.imshow(mask,cmap='gray')
```

```python
masked_image = np.copy(image)
masked_image[mask!=0] = [0,0,0]
plt.imshow(masked_image)
```

```python
# Load in a background image, and convert it to RGB 
background_image = cv2.imread('/Users/taniagretton/miniconda3/CHALLENGES/sky.jpg.webp')
background_image = cv2.cvtColor(background_image,cv2.COLOR_BGR2RGB)
crop_background = background_image[0:360,0:640]
crop_background[mask == 0] = [0,0,0]
plt.imshow(crop_background)
```

```python
complete_image = masked_image + crop_background
```

```python
plt.imshow(complete_image)
```
